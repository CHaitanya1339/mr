```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity

# Convert bag of words to TF-IDF  
tfidf_transformer = TfidfTransformer()
tfidf = tfidf_transformer.fit_transform(vec_matrix)

# Apply LSA or LSI
lsa = TruncatedSVD(n_components=20, algorithm='arpack')
lsa.fit(tfidf)

original_data = tfidf.toarray()
reconstructed_data = lsa.inverse_transform(lsa.transform(original_data))
reconstruction_error = np.sqrt(np.mean((original_data - reconstructed_data)**2))
print("Reconstruction Error:",reconstruction_error)
```
